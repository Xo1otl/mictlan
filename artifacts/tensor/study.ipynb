## 共変ベクトルの例

$$
V を内積を用いた座標系で表すことを考える \\
内積の座標系では各成分を、"基底 \cdot ベクトル"で表すと定義する \\
A_1 = e^1 \cdot V とする \\
e^1 ではなくe^2 = 2 e^1 であるような新しい基底 e^2 で表すことを考える \\
A_1' = e^2 \cdot V となる \\
A_1' = e^2 \cdot V = 2 e^1 \cdot V = 2 A_1 \\
よって新しい成分は、基底の長さに比例する
$$

## 反変ベクトルの例

$$
Vを直交座標系で表すことを考える \\
直交座標系では各成分を、それぞれの成分の方向にどれだけ移動することで目的のベクトルの終端に到達するかによって表すと定義する \\
Vの e_1 成分を A_1 とする \\
この A_1 の値が一意に定まることは自明。 \\
e^1 ではなくe^2 = 2 e^1 であるような新しい基底 e^2 で表すことを考える \\
A_1 e_1 = A_1' e_2 = 2 A_1' e_1 \\
\therefore A_1' = 1/2 A_1 \\
よって新しい成分は、基底の長さに反比例する
$$
