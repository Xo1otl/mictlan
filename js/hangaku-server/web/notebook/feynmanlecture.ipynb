# ファインマン名言メモ

ファインマン先生の真骨頂が出てくる章はいくらもあるが、「本の表紙で中身を読む」の章が、ちょっとお勧めだ。

そのころ先生はカリフォルニア工科大学にいたのだが、カリフォルニアの算数の教科書づくりにかかわってほしいという依頼をうける。当時のアメリカはソ連に宇宙開発の先を越され、数学教育に対する猛然とした反省の嵐が吹いていた。そこでニューマスとよばれる新数学を導入して一挙に汚名を返上しようとし、先生にその教科書づくりを期待したものらしい。

そこで先生はそのころに刊行されたか、準備されつつあった教科書を片っ端から読まされる。先生は地下室に教科書用の書棚をつくって、次から次へと批評を始める。ところがこれがひどいもので、科学や数学を教えるといいながら、考え方をひとつも書いてない。たとえば「７進法であらわされた次の数列を５進法に書きかえよ」といったふうの、何にもならないことばかりを教えようとしているのである。

なかでマシなのは、ぜんまいオモチャの絵、自動車の絵、自転車に乗った子供の絵などで始まるもので、「これは何の力で走るのだろう？」と書いてある。うん、これはいいぞと先生は期待する。けれども次のページを見ると、どの絵にも「これはエネルギーによって動いているのです」とある。ファインマン先生は怒り出す。これでは何もわからない。

先生ならどうするか。こんな例示を本書は示している。

オヤジが子供に話している。「おまえは、何がモノを動かしていると思っているんだ？」「わかんない」「太陽が照ってるからなんだぞ」。そこで子供が反論する「ちがうよ、オモチャはぜんまいを巻くんだ」。オヤジが聞く「じゃあ、どうやってぜんまいは巻かれたんだ？」「ぼくが巻いたんだ」「それでおまえはどうして動けるんだ？」「いろいろ食べてるからね」。こうしてオヤジが得々として言う、「その食べ物はもともと太陽が照っているから育つんだ。だから太陽が照ってるからみんなが動くんだ」。

ファインマンは「意味をつくりだそうとすること」が「考えること」であり、そのことを伝えようとすることが「教育」や「学習」であると確信する。「エネルギーによる」なんて説明は「宇宙はワカリクセスによって動いている」とか「それはピリトミルのせいなんだ」と言っているのと同じで、科学にはならないというわけなのである。
