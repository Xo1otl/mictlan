### 2.3 Coupled-Mode Equations

The coupled-mode theory provides a powerful tool for the theoretical analysis of problems where distributed coupling between optical waves is involved [2,1]-[2.6], [2.10],[2.12], [2.18],[2.20]. A similar theory employed to analyze NLO interactions was given in the seminal paper by Armstrong et al. [2,21]. In this section, we derive the coupled-mode equations in such a form that they are convenient for analysis of nonlinear-optic interactions in waveguides.

Consider a waveguide described by a relative permittivity distribution $\epsilon(x,y)$. At first, we assume that the waveguide materials do not exhibit nonlinearity. A guided mode $m$ of frequency $\omega$ propagating in the waveguide can be written as

$$
E^{(0)}(x,y,z) = E_m(x,y)\exp(-j\beta_mz), \tag{2.52a}
$$

$$
H^{(0)}(x,y,z) = H_m(x,y)\exp(-j\beta_mz), \tag{2.52b}
$$

where $\{E_m, H_m\}$ is the normalized mode profile and $\beta_m$ the propagation constant. Then $\{E^{(0)}, H^{(0)}\}$ satisfies Maxwell equations

$$
\nabla \times E^{(0)} = -j\omega\mu_0H^{(0)}, \quad \nabla \times H^{(0)} = +j\omega\epsilon_0\epsilon E^{(0)}. \tag{2.53}
$$

If the waveguide has optical nonlinearity, nonlinear polarization is induced. We represent by $P$ the component with the same frequency $\omega$ of the nonlinear polarization. We can also include in $P$ additional linear polarization of frequency $\omega$, which may result from modification of the waveguide structure made for the purpose of phase matching, etc. Then, the $\omega$ field in the nonlinear waveguide, $\{E, H\}$ satisfies Maxwell equations

$$
\nabla \times E = -j\omega\mu_0H, \quad \nabla \times H = +j\omega (\epsilon_0\epsilon E + P). \tag{2.54}
$$

From (2.53), (2.54) and vector equations, we obtain

$$
\nabla (E \times H^{(0)^*} + E^{(0)^*} \times H) = -j\omega E^{(0)^*}P. \tag{2.55}
$$

Integration of (2.55) in a volume, parallel to the $xy$ plane, with infinitely small thickness and infinite area yields

$$
\iint \frac {\partial}{\partial z} \left[ E_t \times H_t^{(0)^*} + E_t^{(0)^*} \times H_t \right]_z dxdy = -j\omega \iint {E}^{(0)^*} Pdxdy, \tag{2.56}
$$

where subscript $i$ denotes transverse components. We expand $\{E, H\}$ by the normal modes $\{E_m, H_m\}$ in the original canonical waveguide as

$$
E_{t}(x,y,z) = \sum_{m} A_{m}(z) E_{tm}(x,y) \exp(-j\beta_m z), \tag{2.57a}
$$

$$
H_{t}(x,y,z) = \sum_{m} A_{m}(z) H_{tm}(x,y) \exp(-j\beta_m z), \tag{2.57b}
$$

and substitute (2.52) and (2.57) into (2.56). Then, making use of the orthonormal relation (2.12a), we obtain

$$
\frac{d}{dz} A_{m}(z) = -j\left(\frac{\omega}{4}\right) \iint E_{m}^{*}(x,y) \exp(+j\beta_m z) P(x,y,z) dxdy. \tag{2.58}
$$

This is the fundamental formalism of the coupled-mode theory.

NLO interactions in waveguides are interpreted in terms of distributed coupling between waveguide modes. In many cases, waves of different frequencies are involved. Although it is not explicitly denoted, (2.58) represents a set of differential equations for each frequency. In the following, superscripts will be used to specify the frequency. In addition, (2.58) is valid for each mode $m$. The right hand side of (2.58) includes nonlinear terms of mode amplitudes and may include mode amplitudes at other frequencies, since the polarization is a nonlinear function of the total electric field. Thus (2.58) interprets nonlinear interactions including those across frequencies. If only one mode for each frequency is involved in the coupling, the mode subscript $m$ and $\Sigma$ in (2.57) can be omitted.

To illustrate the specific form of the coupled-mode equations, let us consider guided-mode second-harmonic generation (SHG) as an example. A detailed analysis will be given in the next chapter. Substituting (2.57a) into (2.43) and substituting the results into (2.58) yield

$$
\frac{d}{dz} A^{\omega}(z) = -j\kappa^{*}[A^{\omega}(z)]^{*} A^{2\omega}(z) \exp(-j(2\Delta)z), \tag{2.59a}
$$

$$
\frac{d}{dz} A^{2\omega}(z) = -j\kappa[A^{\omega}(z)]^{2} \exp(+j(2\Delta)z), \tag{2.59b}
$$

$$
2\Delta = \beta^{2\omega} - 2\beta^{\omega} \tag{2.60}
$$

$$
\kappa = \frac{2\omega \epsilon_0}{4} \iint [E^{2\omega}(x,y)]^{*} d(x,y)[E^{\omega}(x,y)]^{2} dxdy. \tag{2.61}
$$

The mode label $m$ has been omitted. Equations (2.59) are the nonlinear coupled-mode equations, which describe the spatial evolution of the amplitudes of the fundamental wave $A^{\omega}$ and that of the second-harmonic wave $A^{2\omega}$. The parameter $k$ defined by (2.61) is called nonlinear coupling coefficient, and it is proportional to an overlap integral of the profiles of optical modes and the nonlinearity.

Consider a case where a pump wave of frequency $\omega$ is incident at $z=0$ in the nonlinear waveguide and propagates to $z=L$. If we assume that the interaction is weak, then the pump amplitude is nearly constant so that we can put $A^{\omega}(z)=A_{0}$. This approximation is called no pump depletion approximation (NPDA). Then, (2.59b) can easily be integrated with boundary condition $A^{2\omega}(0)=0$ to yield

$$
A^{2\omega}(z) = -jkA_0^2z\exp(j\Delta z)(\sin(\Delta z)/\Delta z). \tag{2.62}
$$

Since the mode fields are normalized, the SHG conversion efficiency can be written as

$$
\eta = |A^{2\omega}(L)|^2/|A_0|^2 = |\kappa|^2P_0L^2(\sin(\Delta L)/\Delta L)^2. \tag{2.63}
$$

where $P_0 = |A_0|^2$ is the incident pump power.

### Phase Matching

#### 2.4.1 Coherence Length and Phase Matching

Figure 11 illustrates one-dimensional traveling-wave SHG described by (2.59)-(2.63). The second-harmonic component of the NLO polarization $\bm{P}^{2\omega}$ at each point in the medium gives rise to generation of wavelet at frequency $2\omega$ which propagates in the medium. The total SHG results from accumulation of the harmonic wavelets from all points where the pump wave propagates.

The phase distribution of $\bm{P}^{2\omega}$ along the $z$ axis is described by a phase constant $2\bm{\beta}^{\omega}$, since $\bm{P}^{2\omega}$ is proportional to square of $E^{\omega}$ propagating with phase constant $\bm{\beta}^{\omega}$. The wavelets, on the other hand, propagate with phase constant $\bm{\beta}^{2\omega}$. The parameter $2\Delta$ defined by (2.60) may be written as

$$
2\Delta = \bm{\beta}^{2\omega} - 2\bm{\beta}^{\omega} = \frac{4\pi}{\lambda}(N^{2\omega} - N^{\omega}) = \frac{\pi}{L_{c}}, \tag{2.64}
$$

$$
L_{c} = \frac{\lambda}{4(N^{2\omega} - N^{\omega})}, \tag{2.65}
$$

where $N^{\omega}$ and $N^{2\omega}$ are the mode indexes for the pump and harmonic waves, respectively, and $\lambda$ is the wavelength of the pump wave in vacuum. The length $L_{c}$ defined by (65) is called coherence length. If $\bm{\beta}^{2\omega} = 2\bm{\beta}^{\omega}$ ($N^{2\omega} = N^{\omega}$) so that $\Delta = 0$ and $L_{c} \rightarrow \infty$, all the harmonic wavelets are in phase and are superimposed constructively. This condition is called phase matching. Under phase matching, the harmonic amplitude grows in proportion to $z$, as shown in Fig. 11 (a), and therefore, efficient SHG takes place. Equation (63) shows that the SHG efficiency in NPDA under phase matching is proportional to the pump power, the square of the coupling coefficient, and the square of the interaction length.

If $\bm{\beta}^{2\omega} \neq 2\bm{\beta}^{\omega}$ ($N^{2\omega} \neq N^{\omega}$) so that $\Delta \neq 0$, the harmonic wavelets are dephased from each other and are superimposed destructively. The harmonic amplitude ripples with a period corresponding to two times coherence length, as shown in Fig. 11 (b), but does not grow with propagation. The constructive superposition takes place only with the wavelets generated within the coherence length, and the wavelets generated in the next coherence length cancel out the previous wavelets. Therefore, if the phase mismatch $\Delta$ is large, SHG is not substantially accomplished.

### 2.4.4 Quasi-Phase Matching

Consider a structure where the sign of the nonlinear coefficient $d$ is inverted with a period $\Lambda$ in the direction of wave propagation ($z$ direction). For such periodic (grating) structures, $d$ can be expressed in a form of Fourier series

$$
d(x,y,z) = \sum_{q} d_q(x,y) \exp(-jqKz), \quad K=2\pi/\Lambda. \tag{2.66}
$$

Accordingly, the right hand sides of the coupled-mode equations are expressed by a summation of the spatial harmonic terms. Only one term of them can contribute to NLO interaction, the other terms do not contribute substantially since they oscillate rapidly with $z$, and therefore, we can omit them. Then the coupled-mode equations are given by (2.59) with (2.60) replaced by

$$
2\Delta = \beta^{2\omega} - (2\beta^{\omega} + qK). \tag{2.67}
$$

This result shows that, when $\Delta = 0$ holds with (2.67), the harmonic wave propagation phase matches one of the spatial harmonics of the nonlinear polarization of harmonic wavelength, and efficient SHG takes place. This is called quasi-phase matching (QPM) or grating phase matching. The wave vector diagram is shown in Fig. 2.13 (c). QPM may be considered compensation for the wave-vector difference between nonlinear polarization and harmonic waves by the grating vector $K$. The device configuration will be shown in the next chapter.

The period for first-order QPM coincides with two times the coherence length given by (2.65). This relation can be used to explain QPM in an alternative way. After constructive NLO interaction within the coherence length, the sign of $d$ is inverted within the next coherence length in order to convert the otherwise destructive contribution to constructive one, and the same process is repeated periodically to accumulate substantially the NLO interaction [2.21]. The above spatial harmonics interpretation is more general and useful for cases of high efficiency where the interaction length is much larger than the coherence length. As we will see in detail in the next chapter, QPM can be accomplished not only with periodic inversion of nonlinear permittivity but also with more general periodic modulation of nonlinear and/or linear permittivities.

QPM was first proposed in bulk configurations [2.21], but is now widely applied in both bulk and waveguide configurations. Although QPM requires a periodic structure, it offers a number of advantages. One of them is that matching can be accomplished at any wavelength in the transparency region of the medium by only appropriate design of the QPM period. This is a very important advantage from the device implementation point of view. Matching can be accomplished with arbitrary combinations of mode and polarization. This means that the largest element of the SHG tensor can be used. The largest element of many NLO crystals is a diagonal element, which cannot be used for birefringence phase matching. Use of the largest tensor element, along with use of fundamental modes for pump and harmonic waves of large mode overlap, allows a large coupling coefficient, and therefore, high efficiency. The possibility of using chirped or spatially-modulated grating structures provides additional design flexibility.
QPM is therefore the most versatile phase matching technique for implementation of paractical devices, only if the technique to form the periodic structure is found.

# Expression of Waveguide with Grating

Consider a channel waveguide having a uniform cross section, as a canonical structure. The channel is assumed to be parallel to one of the optical principal axes of the waveguide material, and the axis is taken as the $z$ axis of the coordinate system. The guide is described by the relative permittivity profile $\mathbf{e}(x, y)$. We assume that the guide supports at least one guided mode for both fundamental (pump) and second-harmonic (SH) waves.

Figure 3.1 illustrates a channel waveguide with a grating of period $A$ for quasi-phase matching (QPM). The device can be fabricated by giving a periodic modulation of optical constants (permittivity and/or SHG coefficient) to the waveguide: the structure can be described by the distributions of the permittivity and the SHG coefficient. Let $\Delta \mathbf{e}(x,y,z)$ be the periodic modulation in relative permittivity, then $\Delta \mathbf{e}$ can be expressed in the form of a Fourier series:

$$
\Delta \epsilon(x,y,z) = \sum_{q} \Delta \epsilon_q(x,y) \exp(-jqKz), \quad K=2\pi/\Lambda, \tag{3.1}
$$

where $\Delta \mathbf{e}_q$ is the amplitude of the $q$-th order Fourier component, and $K$ is the spatial frequency of the grating, i.e., the magnitude of the grating vector. The distribution of the SHG coefficient with periodic modulation can be written as

$$
d(x,y,z) = \sum_{q} d_q(x,y) \exp(-jqKz), \quad K=2\pi/\Lambda. \tag{3.2}
$$

# Nonlinear Coupled-Mode Equations

We derive nonlinear coupled-mode equations to describe SHG in waveguides with a grating. We use $\omega$ and $\lambda$ to denote the frequency and vacuum wavelength of the fundamental (pump) wave, respectively, and $2\omega$ and $\lambda/2$ for the second-harmonic (SH) wave.

The fundamental formula for deriving the coupled-mode equations is given by (2.58). Although (2.57) and (2.58) are written for the pump waves of frequency $\omega$, we have another set of equations for the SH waves, and they are given by replacing $\omega$ by $2\omega$ in (2.57) and (2.58). The polarization $P(x, y, z)$ in the present SHG device can be written as

$$
P^{\omega} = P_L^{\omega} + P_{NL}^{\omega} = \varepsilon_0 A\varepsilon^\omega E^\omega + 2\varepsilon_0 dE^{\omega*} E^{2\omega}, \tag{3.5a}
$$

$$
P^{2\omega} = P_L^{2\omega} + P_{NL}^{2\omega} = \varepsilon_0 A\varepsilon^{2\omega} E^{\omega} + \varepsilon_0 dE^{\omega*} E^{\omega}, \tag{3.5b}
$$

where $P_L$ is the linear polarization due to the grating permittivity, $P_{NL}$ the nonlinear polarization due to the SHG coefficient, and use has been made of (2.43).

Here we assume that there is no substantial coupling between modes of same frequency, but a mode of pump frequency couples with a mode of SH frequency through the quasi-phase matching (QPM) of the fundamental diffraction order ($q=1$). The condition for QPM (including a slight phase mismatch) is given by

$$
2\beta^{\omega} + K \simeq \beta^{2\omega} \quad \text{or} \quad \Lambda \simeq (\lambda/2)/(N^{2\omega} - N^{\omega}), \tag{3.6}
$$

where $N^{\omega}$ and $N^{2\omega}$ are the mode indexes of the pump and SH waves, respectively. We also neglect the guided-mode to radiation-mode coupling.
