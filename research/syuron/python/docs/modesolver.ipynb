**異方性材料の電磁波解析に関する詳細な解説**

本稿では、**横方向異方性材料**（transverse anisotropic materials）について考察します。特に、材料の主要軸の一つが導波路の方向（通常は z 軸）に沿っている場合を想定します。この仮定の下で、誘電率テンソルは以下の形をとります。

---

**式 (1): 誘電率テンソル**

$$
{\boldsymbol{\epsilon}} \equiv \epsilon_{0} \begin{bmatrix}
\epsilon_{xx} & \epsilon_{xy} & 0 \\
\epsilon_{yx} & \epsilon_{yy} & 0 \\
0 & 0 & \epsilon_{zz}
\end{bmatrix}.
$$

---

ここで、$\epsilon_{0}$ は真空の誘電率であり、$\epsilon_{ij}$ は材料の相対誘電率のテンソル成分を表します。

**座標系の回転と有限差分法の適用**

一般的には、適切な回転座標系を選ぶことで、この誘電率テンソルを対角化することが可能です。しかし、以下の理由から、そのような回転座標系で有限差分法を適用することは不便です。

1. **グリッドとの整合性の問題**: 座標系を回転させると、導波路の境界が有限差分グリッドに対して斜めになり、計算が複雑になります。

2. **複数の異方性領域の存在**: 導波路内に異なる方向に異方性を持つ複数の材料が存在する場合、全ての材料の誘電率テンソルを同時に対角化することはできません。

3. **磁気光学材料の場合**: 磁気光学材料では、誘電率テンソルに虚数の非対角成分が追加され、単純な回転では対角化が不可能になります。

**横方向磁場成分によるモードの記述**

誘電率テンソルが上記の形（式 (1)）の場合、光導波路の**ベクトル固有モード**は横方向の磁場成分 $H_x$ と $H_y$ の2つで完全に記述できます。他の電磁場成分はこれらから導出可能です [参考文献8]。

**マクスウェルの方程式からの固有値方程式の導出**

マクスウェルの回転（カール）方程式を組み合わせることで、横方向磁場に関する固有値方程式を得ます。

---

**式 (2): 磁場に関する固有値方程式**

$$
\nabla \times ({\boldsymbol{\epsilon}}^{-1} \nabla \times {\bf H}) - \omega^{2} \mu_{0} {\bf H} = 0.
$$

---

ここで、

- $\nabla \times$ は回転（カール）演算子、
- ${\boldsymbol{\epsilon}}^{-1}$ は誘電率テンソルの逆行列、
- $\omega$ は角周波数、
- $\mu_{0}$ は真空の透磁率、
- ${\bf H}$ は磁場ベクトル、

を表します。また、時間依存性として $e^{j\omega t}$ を仮定しています。

**縦方向磁場成分の表現**

z 方向の依存性を $e^{-j\beta z}$ と仮定すると、磁場の発散条件 $\nabla \cdot {\bf H} = 0$ から、縦方向成分 $H_z$ を横方向成分で表すことができます。

---

**式 (3): 縦方向磁場成分 $H_z$ の表現**

$$
H_{z} = \frac{1}{j\beta} \left( \frac{\partial H_{x}}{\partial x} + \frac{\partial H_{y}}{\partial y} \right).
$$

---

ここで、$\beta$ は伝搬定数です。

**横方向磁場成分のみを用いた固有値方程式の導出**

式 (3) を用いて、式 (2) を横方向成分 $H_x$ と $H_y$ のみで表現します。計算と代数的な整理を経て、以下の連立した固有値方程式を得ます。

---

**式 (4a): $H_x$ に関する固有値方程式**

$$
\begin{aligned}
&\frac{\partial^{2} H_{x}}{\partial x^{2}} + \frac{\epsilon_{yy}}{\epsilon_{zz}} \frac{\partial^{2} H_{x}}{\partial y^{2}} + \frac{\epsilon_{yx}}{\epsilon_{zz}} \frac{\partial^{2} H_{x}}{\partial y \partial x} + \left(1 - \frac{\epsilon_{yy}}{\epsilon_{zz}}\right) \frac{\partial^{2} H_{y}}{\partial x \partial y} \\
&\quad - \frac{\epsilon_{yx}}{\epsilon_{zz}} \frac{\partial^{2} H_{y}}{\partial x^{2}} + k^{2} (\epsilon_{yy} H_{x} - \epsilon_{yx} H_{y}) = \beta^{2} H_{x}.
\end{aligned}
$$

---

**式 (4b): $H_y$ に関する固有値方程式**

$$
\begin{aligned}
&\frac{\partial^{2} H_{y}}{\partial y^{2}} + \frac{\epsilon_{xx}}{\epsilon_{zz}} \frac{\partial^{2} H_{y}}{\partial x^{2}} + \frac{\epsilon_{xy}}{\epsilon_{zz}} \frac{\partial^{2} H_{y}}{\partial x \partial y} + \left(1 - \frac{\epsilon_{xx}}{\epsilon_{zz}}\right) \frac{\partial^{2} H_{x}}{\partial y \partial x} \\
&\quad - \frac{\epsilon_{xy}}{\epsilon_{zz}} \frac{\partial^{2} H_{x}}{\partial y^{2}} + k^{2} (\epsilon_{xx} H_{y} - \epsilon_{xy} H_{x}) = \beta^{2} H_{y}.
\end{aligned}
$$

---

ここで、$k^{2} = \omega^{2} \mu_{0} \epsilon_{0}$ です。

**式 (4) の各項の物理的意味**

- **二階偏微分項**: 空間的な変化、つまり磁場の分布の変化率を表します。
- **誘電率の比**: $\epsilon_{ij}/\epsilon_{zz}$ の形で現れ、異方性の影響を示します。
- **混合偏微分項**: 異方性材料では、$x$ と $y$ の成分が結合するため、混合項が現れます。
- **質量項（$k^{2}$ を含む項）**: 材料の誘電特性と磁場の相互作用を示します。

**電束密度 $\mathbf{D}$ の計算**

磁場の全成分が求まれば、マクスウェルの方程式 $\nabla \times \mathbf{H} = j\omega \mathbf{D}$ を用いて、電束密度 $\mathbf{D}$ を計算できます。

---

**式 (5): 電束密度の成分**

$$
\begin{aligned}
D_{x} &= -\frac{1}{\omega \beta} \left( \frac{\partial^{2} H_{x}}{\partial y \partial x} + \frac{\partial^{2} H_{y}}{\partial y^{2}} \right) + \frac{\beta}{\omega} H_{y}, \\
D_{y} &= \frac{1}{\omega \beta} \left( \frac{\partial^{2} H_{y}}{\partial x \partial y} + \frac{\partial^{2} H_{x}}{\partial x^{2}} \right) - \frac{\beta}{\omega} H_{x}, \\
D_{z} &= \frac{j}{\omega} \left( \frac{\partial H_{x}}{\partial y} - \frac{\partial H_{y}}{\partial x} \right).
\end{aligned}
$$

---

**電場 $\mathbf{E}$ の計算**

電束密度 $\mathbf{D}$ から電場 $\mathbf{E}$ を求めるには、誘電率テンソルの逆行列を適用します。

---

**式 (6): 横方向電場成分**

$$
\begin{bmatrix}
E_{x} \\
E_{y}
\end{bmatrix}
= \frac{1}{\epsilon_{0} (\epsilon_{xx} \epsilon_{yy} - \epsilon_{xy} \epsilon_{yx})}
\begin{bmatrix}
\epsilon_{yy} & -\epsilon_{xy} \\
-\epsilon_{yx} & \epsilon_{xx}
\end{bmatrix}
\begin{bmatrix}
D_{x} \\
D_{y}
\end{bmatrix}.
$$

---

**式 (7): 縦方向電場成分**

$$
E_{z} = \frac{1}{\epsilon_{0} \epsilon_{zz}} D_{z} = \frac{j}{\omega \epsilon_{0} \epsilon_{zz}} \left( \frac{\partial H_{x}}{\partial y} - \frac{\partial H_{y}}{\partial x} \right).
$$

---

**まとめ**

以上の手順により、異方性材料の光導波路における電磁場の解析が可能となります。具体的には、横方向の磁場成分 $H_x$ と $H_y$ を求めることで、他の全ての電磁場成分を導出できます。このアプローチは、異方性材料や磁気光学材料を含む複雑な導波路構造の解析において特に有用です。

- **有限差分法の適用**: 導波路の境界がグリッドに整合しているため、計算効率が向上します。
- **複数の異方性領域への対応**: 座標系の回転を行わないため、異なる異方性材料を含む場合でも一貫した解析が可能です。
- **磁気光学効果の考慮**: 誘電率テンソルの非対角成分が虚数であっても、この方法で解析できます。

以上、異方性材料の電磁場解析に関する詳細な解説を行いました。各式の導出過程や物理的意味を理解することで、異方性材料を含む光導波路の設計や解析に役立てることができます。

**誘電率テンソルの詳細な解説**

**1. 誘電率テンソルとは**

誘電率テンソル $\boldsymbol{\epsilon}$ は、材料内での電場 $\mathbf{E}$ と電束密度 $\mathbf{D}$ の関係を記述するためのテンソル（行列）です。一般的に、これらの関係は以下の式で表されます。

$$
\mathbf{D} = \epsilon_0 \boldsymbol{\epsilon} \mathbf{E}
$$

ここで、

- $\mathbf{D}$：電束密度（単位：クーロン毎平方メートル, C/m²）
- $\epsilon_0$：真空の誘電率（約 $8.854 \times 10^{-12}$ F/m）
- $\boldsymbol{\epsilon}$：相対誘電率のテンソル（無次元）
- $\mathbf{E}$：電場（単位：ボルト毎メートル, V/m）

材料が**等方性**である場合、$\boldsymbol{\epsilon}$ はスカラー値（単一の数値）となり、$\mathbf{D} = \epsilon_0 \epsilon_r \mathbf{E}$ となります。ここで、$\epsilon_r$ は相対誘電率です。しかし、材料が**異方性**である場合、$\boldsymbol{\epsilon}$ はテンソルとなり、電場の各成分と電束密度の各成分が相互に関連します。

**2. 異方性材料と横方向異方性材料**

**異方性材料**とは、その物理的特性（この場合は誘電率）が方向によって異なる材料を指します。つまり、電場の方向が異なれば、材料の応答（電束密度）も異なります。

**横方向異方性材料（transverse anisotropic materials）**は、特定の軸（通常は z 軸）に対して対称性を持ち、他の方向には異方性を示す材料です。この場合、z 軸方向には等方的であり、x-y 平面内で異方性を示します。

**3. 材料の主要軸と誘電率テンソル**

材料には、その物理的特性が最も単純な形で表される**主要軸（principal axes）**があります。これらの軸に沿った座標系では、誘電率テンソルは対角化され、非対角成分がゼロになります。しかし、本稿では、材料の主要軸の一つが導波路の方向（z 軸）に沿っていると仮定しています。

この仮定の下で、誘電率テンソルは以下の形をとります。

---

**式 (1): 誘電率テンソル**

$$
\boldsymbol{\epsilon} = \epsilon_0 \begin{bmatrix}
\epsilon_{xx} & \epsilon_{xy} & 0 \\
\epsilon_{yx} & \epsilon_{yy} & 0 \\
0 & 0 & \epsilon_{zz}
\end{bmatrix}
$$

---

ここで、

- $\epsilon_{xx}, \epsilon_{yy}, \epsilon_{zz}$：対角成分であり、それぞれ x, y, z 軸方向の相対誘電率を表します。
- $\epsilon_{xy}, \epsilon_{yx}$：非対角成分であり、x と y の電場成分が相互に影響し合うことを示します。
- 0：x-z および y-z 平面では異方性がないことを示します。

**4. 誘電率テンソルの成分の物理的意味**

- **対角成分（$\epsilon_{xx}, \epsilon_{yy}, \epsilon_{zz}$）**：各軸方向の電場に対する電束密度の直接的な応答を示します。たとえば、$\epsilon_{xx}$ は x 方向の電場が x 方向の電束密度にどのように影響するかを表します。
  
- **非対角成分（$\epsilon_{xy}, \epsilon_{yx}$）**：一つの軸方向の電場が別の軸方向の電束密度に影響を与えることを示します。たとえば、$\epsilon_{xy}$ は x 方向の電場が y 方向の電束密度に影響を与える度合いを表します。

**5. テンソルの対称性と非対称性**

- **対称テンソル**：$\epsilon_{xy} = \epsilon_{yx}$ の場合、誘電率テンソルは対称であり、材料は**損失のない媒質**を表します。この対称性は、エネルギーの保存則や材料の双極子性から導かれます。

- **非対称テンソル**：$\epsilon_{xy} \neq \epsilon_{yx}$ の場合、テンソルは非対称であり、材料は**磁気光学効果**などの非双極子性を示す可能性があります。特に、$\epsilon_{xy} = -\epsilon_{yx}$ の場合、テンソルは**反対称**となり、磁場の影響を受けた材料（たとえばファラデー効果を示す材料）を表します。

**6. 誘電率テンソルの対角化とその困難性**

一般的には、適切な座標系への回転を行うことで、誘電率テンソルを対角化し、非対角成分を消去することが可能です。しかし、以下の理由により、この方法が実用的でない場合があります。

- **有限差分法との不整合**：座標系を回転すると、導波路の物理的な境界が計算グリッドと一致しなくなり、境界条件の設定や計算が複雑になります。

- **複数の異方性材料の存在**：異なる方向に異方性を持つ複数の材料が存在する場合、単一の回転座標系で全ての誘電率テンソルを同時に対角化することは不可能です。

- **磁気光学材料の扱い**：磁気光学材料では、誘電率テンソルに虚数の非対角成分が含まれます。この場合、単純な回転ではテンソルを対角化できません。

**7. 誘電率テンソルと光の伝搬**

異方性材料中での光の伝搬は、誘電率テンソルの形状に大きく依存します。特に、非対角成分の存在は、光の偏光状態や伝搬方向に影響を与えます。

- **複屈折**：材料が異方性である場合、異なる偏光状態の光が異なる速度で伝搬します。これは誘電率テンソルの対角成分の違いによるものです。

- **光の偏光回転**：非対角成分が存在すると、光の偏光面が回転する現象が起こります。これはファラデー効果や旋光性によるものです。

**8. 誘電率テンソルの数学的取り扱い**

誘電率テンソルは、一般的に複素対称行列として扱われます。これは、損失のない媒質ではテンソルがエルミート対称であり、固有値が実数となるためです。

- **固有値と固有ベクトル**：誘電率テンソルの固有値と固有ベクトルを求めることで、材料の主軸系を見つけることができます。

- **対角化の条件**：テンソルを対角化するためには、非対角成分がゼロになるような座標変換が必要です。しかし、物理的な制約や計算上の理由から、これが適切でない場合があります。

**9. 実際の応用例**

- **液晶ディスプレイ**：液晶材料は異方性を示し、その誘電率テンソルは電圧の印加により変化します。これにより、光の透過や偏光状態を制御できます。

- **光ファイバー**：偏波保持ファイバーなどでは、誘電率テンソルの異方性を利用して特定の偏光状態を維持します。

- **磁気光学デバイス**：磁場によって誘電率テンソルの非対角成分が変化し、光の偏光状態を制御するデバイス（例：光アイソレータ、サーキュレータ）に応用されます。

**10. まとめ**

誘電率テンソルは、異方性材料における電場と電束密度の関係を詳細に記述するための重要なツールです。特に、非対角成分の存在は材料の異方性や磁気光学効果を反映しており、光の伝搬や偏光状態に大きな影響を与えます。

本稿で示した誘電率テンソル（式 (1)）は、z 軸方向に沿った対称性を持つ横方向異方性材料をモデル化しています。このような材料では、x-y 平面内での異方性が考慮され、電磁場の解析が複雑になります。

しかし、座標系の回転による対角化が困難な場合でも、非対角成分を含むテンソルのままで解析を進めることが可能です。これにより、複雑な材料特性や構造を持つ導波路の設計・解析が可能となります。
